```python
import pandas as pd
import numpy as np
import tensorflow as tf
import keras
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree
```

```python
train = pd.read_csv("dataset_train_set.csv")
test = pd.read_csv("dataset_dev_set.csv")
```

```python
X_train = train.drop('target',axis=1)
y_train = train['target']
X_test = test.drop('target',axis=1)
y_test = test['target']
```

```python
drop_list = ['aPosLS','hPosLS','hGD5','aSpent','aPtsPG','hSPG','hGCPG','aPos','aSTPG','aGSPG','hSTPG','aSPG','aGD7','hPos','aGD5','hPtsPG','hSpent','aPTSPG7','hGD7','aGCPG','aPTSPG5','hPTSPG7']
```

```python
X_train.drop(drop_list,inplace=True,axis=1)
X_test.drop(drop_list,inplace=True,axis=1)
```

```python
X_train.head()
```

```python
clf_gini = DecisionTreeClassifier(criterion = "gini", random_state = 100,
                               max_depth=3, min_samples_leaf=5)
clf_gini.fit(X_train, y_train)
```

```python
y_pred = clf_gini.predict(X_test)

print("Accuracy is {}".format(accuracy_score(y_test,y_pred)*100))
```

```python
import eli5
from eli5.sklearn import PermutationImportance
```

```python
perm = PermutationImportance(clf_gini, random_state=1).fit(X_test, y_test)
a = eli5.show_weights(perm, feature_names = X_test.columns.tolist(), top=36)
html = a.data
with open('permutation_importance.html', 'w') as f:
    f.write(html)
a
```
