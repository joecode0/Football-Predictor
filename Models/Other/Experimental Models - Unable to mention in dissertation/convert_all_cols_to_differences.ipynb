```python
import pandas as pd
df = pd.read_csv("main_database_for_models.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
def convert_cols_to_difference(row,col1,col2):
    val1 = float(row[col1])
    val2 = float(row[col2])
    new_val = val1-val2
    return round(float(new_val),3)
```

```python
df['PosDiff'] = df.apply(convert_cols_to_difference,args=('hPos','aPos'),axis=1)
df['GSPGDiff'] = df.apply(convert_cols_to_difference,args=('hGSPG','aGSPG'),axis=1)
df['GCPGDiff'] = df.apply(convert_cols_to_difference,args=('hGCPG','aGCPG'),axis=1)
df['CSPGDiff'] = df.apply(convert_cols_to_difference,args=('hCSPG','aCSPG'),axis=1)
df['PtsPGDiff'] = df.apply(convert_cols_to_difference,args=('hPtsPG','aPtsPG'),axis=1)
# df['ShotDiff'] = df.apply(convert_cols_to_difference,args=('hs2','as2'),axis=1)
df['SPGDiff'] = df.apply(convert_cols_to_difference,args=('hSPG','aSPG'),axis=1)
df['STPGDiff'] = df.apply(convert_cols_to_difference,args=('hSTPG','aSTPG'),axis=1)
df['S%Diff'] = df.apply(convert_cols_to_difference,args=('hS%','aS%'),axis=1)
df['PTSPG3Diff'] = df.apply(convert_cols_to_difference,args=('hPTSPG3','aPTSPG3'),axis=1)
df['PTSPG5Diff'] = df.apply(convert_cols_to_difference,args=('hPTSPG5','aPTSPG5'),axis=1)
df['PTSPG7Diff'] = df.apply(convert_cols_to_difference,args=('hPTSPG7','aPTSPG7'),axis=1)
df['GD3Diff'] = df.apply(convert_cols_to_difference,args=('hGD3','aGD3'),axis=1)
df['GD5Diff'] = df.apply(convert_cols_to_difference,args=('hGD5','aGD5'),axis=1)
df['GD7Diff'] = df.apply(convert_cols_to_difference,args=('hGD7','aGD7'),axis=1)
df['PosLSDiff'] = df.apply(convert_cols_to_difference,args=('hPosLS','aPosLS'),axis=1)
df['SpentDiff'] = df.apply(convert_cols_to_difference,args=('hSpent','aSpent'),axis=1)
df['IncomeDiff'] = df.apply(convert_cols_to_difference,args=('hIncome','aIncome'),axis=1)
df['NetDiff'] = df.apply(convert_cols_to_difference,args=('hNet','aNet'),axis=1)
```

```python
df.drop(['hPos', 'hGSPG', 'hGCPG', 'hCSPG', 'hPtsPG', 'aPos', 'aGSPG',
       'aGCPG', 'aCSPG', 'aPtsPG', 'hSPG',
       'hSTPG', 'aSPG', 'aSTPG', 'hS%', 'aS%', 'hPTSPG3', 'hPTSPG5', 'hPTSPG7',
       'hGD3', 'hGD5', 'hGD7', 'aPTSPG3', 'aPTSPG5', 'aPTSPG7', 'aGD3', 'aGD5',
       'aGD7', 'hPosLS', 'aPosLS', 'hSpent', 'hIncome', 'hNet', 'aSpent',
       'aIncome', 'aNet'],inplace=True,axis=1)
```
