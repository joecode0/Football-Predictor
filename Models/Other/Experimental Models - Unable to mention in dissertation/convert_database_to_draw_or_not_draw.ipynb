```python
import pandas as pd
df = pd.read_csv("main_database_small_diff.csv")
```

```python
df.head()
```

```python
def target_to_draws_bool(row):
    target = int(row['target'])
    if target == 1:
        return 1
    else:
        return 0
```

```python
df['target'] = df.apply(target_to_draws_bool,axis=1)
```

```python
df.head()
```

```python
df.to_csv("main_database_small_diffs_draws_or_not.csv",index=False)
```
