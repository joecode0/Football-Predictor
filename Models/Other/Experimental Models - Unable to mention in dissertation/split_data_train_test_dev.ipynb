```python
import pandas as pd
df = pd.read_csv("main_database_cleaned_for_models.csv")
df.head()
```

```python
df_season_26 = df.query("season == 26")
df_season_26.reset_index(inplace=True,drop=True)
df_season_26.sort_values(['date','home'],inplace=True,axis=0)
df_rest = df.query("season != 26")
df_rest.reset_index(inplace=True,drop=True)
df_rest.sort_values(['date','home'],inplace=True,axis=0)
```

```python
df_season_26.drop(['date','home'],inplace=True,axis=1)
df_season_26.to_csv("main_database_final_test_set.csv",index=False)
```
