```python
import pandas as pd
import numpy as np
import time
```

```python
df_raw = pd.read_csv("all_seasons_league_table_data_added.csv")
```

```python
df_raw.head()
```

```python
def decide_target_variable(row):
    score = str(row["score"])
    goals = [int(x) for x in score.split("-")]
    if goals[0] > goals[1]:
        return "2"
    elif goals[0] == goals[1]:
        return "1"
    else:
        return "0"

def cleaned_to_model_ready(df_raw):
    df_raw['target'] = df_raw.apply(decide_target_variable,axis=1)
    df_raw.drop(['premid','11id','date','home','score','away','ref','stadium','h_scorers','h_players','a_scorers','a_players'],inplace=True,axis=1)
    return df_raw
```

```python
df_new = cleaned_to_model_ready(df_raw.copy())
df_new.head()
```

```python
df_new.to_csv("initial_dataset_final.csv",index=False)
```
