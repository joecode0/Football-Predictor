```python
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import tensorflow as tf

train = pd.read_csv("dataset_train_set.csv")
dev = pd.read_csv("dataset_dev_set.csv")
test = pd.read_csv("dataset_test_set.csv")
```

```python
X_train = train.drop('target',axis=1)
y_train = train['target']
X_dev = dev.drop('target',axis=1)
y_dev = dev['target']
X_test = test.drop('target',axis=1)
y_test = test['target']
```

```python
X = pd.concat([X_train,X_dev,X_test],axis=0)
y = pd.concat([y_train,y_dev,y_test],axis=0)
X.shape
```

```python
X_train = X[0:4500].copy()
y_train = y[0:4500].copy()
X_test = X[4500:].copy()
y_test = y[4500:].copy()
```

```python
use_list = ['aPosLS','hPosLS','hGD5','aSpent','aPtsPG','hSPG','hGCPG','aPos','aSTPG','aGSPG','hSTPG','aSPG','aGD7','hPos','aGD5','hPtsPG','hSpent','aPTSPG7','hGD7']
X_train = X_train[use_list]
X_test = X_test[use_list]
X_train.head()
```

```python
X_test.shape
```

```python
X_train = tf.keras.utils.normalize(X_train.values,axis=1)
X_test = tf.keras.utils.normalize(X_test.values,axis=1)
```

```python
knn = KNeighborsClassifier(n_neighbors=109)
knn.fit(X_train,y_train)
y_pred = knn.predict(X_test)
print(knn.score(X_test,y_test)*100)
```

```python
df = pd.read_csv("football-data.csv")
df.head()
```

```python
useful_df = df[['Date','HomeTeam','AwayTeam','FTHG','FTAG','BbMxH','BbMxD','BbMxA']].copy()
useful_df.head()
```

```python
def get_result(row):
    h_goals = int(row['FTHG'])
    a_goals = int(row['FTAG'])
    if h_goals > a_goals:
        return 2
    elif h_goals == a_goals:
        return 1
    else:
        return 0
```

```python
useful_df['target'] = useful_df.apply(get_result,axis=1)
```
