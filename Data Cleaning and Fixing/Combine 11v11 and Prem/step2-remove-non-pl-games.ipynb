```python
def test_prem_game(row):
    home = str(row['Home Team'])
    season = int(row['Season'])
    team_seasons = all_teams_seasons_dict.get(home)
    if len(team_seasons) < 1:
        return False
    if season in team_seasons:
        return True
    else:
        #print("{}: {} v {}".format(row['Date'],home,str(row['Away Team'])))
        return False
```

```python
prem['isPrem'] = prem.apply(test_prem_game,axis=1)
```

```python
eleven['isPrem'] = eleven.apply(test_prem_game,axis=1)
```

```python
premNew = prem[prem.isPrem]
elevenNew = eleven[eleven.isPrem]
```
