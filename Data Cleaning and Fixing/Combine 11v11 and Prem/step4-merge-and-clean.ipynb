```python
import pandas as pd
import time
import numpy as np
```

```python
prem = pd.read_csv("prem_finalised.csv")
eleven = pd.read_csv("eleven_finalised.csv")
```

```python
prem.columns = [str(x)+"2" for x in list(prem.columns.values)]
df = pd.concat([prem,eleven],axis=1)
```

```python
df.head()
```

```python
def test_equal(row,col1,col2):
    val1 = str(row[col1])
    val2 = str(row[col2])
    return (val1 == val2)
```

```python
df['home_equal'] = df.apply(test_equal,args=('home','home2'),axis=1)
df['away_equal'] = df.apply(test_equal,args=('away','away2'),axis=1)
df['refs_equal'] = df.apply(test_equal,args=('ref','ref2'),axis=1)
```

```python
df['home_equal'].value_counts()
```

```python
df['away_equal'].value_counts()
```

```python
df['refs_equal'].value_counts()
```

```python
teams_not_equal = df.query("home_equal != True | away_equal != True")
```

```python
teams_not_equal.head()
```

```python
teams_not_equal.to_csv("teams_not_equal.csv",index=False)
```
