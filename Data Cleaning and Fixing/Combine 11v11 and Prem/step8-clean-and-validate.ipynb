```python
import pandas as pd
import numpy as np
import time
```

```python
df = pd.read_csv("fixed_goalscorers_step7.csv")
```

```python
df.info()
```

```python
def check_columns_agree(row,col1,col2):
    val1 = row[col1]
    val2 = row[col2]
    return (val1 == val2)
```

```python
df['home_equal'] = df.apply(check_columns_agree,args=("home","home2"),axis=1)
df['home_equal'].value_counts()
```

```python
df['away_equal'] = df.apply(check_columns_agree,args=("away","away2"),axis=1)
df['away_equal'].value_counts()
```

```python
df['season_equal'] = df.apply(check_columns_agree,args=("season","season2"),axis=1)
df['season_equal'].value_counts()
```

```python
df['date_equal'] = df.apply(check_columns_agree,args=("date","date2"),axis=1)
df['date_equal'].value_counts()
```

```python
dates_disagree = df.query("date_equal == False")
dates_disagree
```

```python
df['date_final'] = df['date']
```

```python
df.head()
```

```python
df_cols_dropped = df.drop(['refs_equal','home2','away2','ref2','ref','season2','date2','date','home_equal','away_equal','season_equal','date_equal'],axis=1)
```

```python
df_cols_dropped.head()
```

```python
df_cols_dropped.columns = ['premid','score','11id','season','home','away','stadium','h_scorers','a_scorers','h_players','a_players','ref','date']
```

```python
df_cols_dropped.head()
```

```python
df_final = df_cols_dropped[['premid','11id','season','date','home','score','away','ref','stadium','h_scorers','a_scorers','h_players','a_players']]
```
