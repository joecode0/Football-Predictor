```python
# Given a league table, need to generate the following statistics:
# Goals per game for 2 teams
# Goals conceded per game
# Clean sheets per game for 2 teams
# Points per game

# So first acquire row for that team
# Then get number of games
# Then divide each of the follow by # of games: GS, GC, CS, Pts

def get_useful_stats(league_table,row):
    homeName = row['home']
    awayName = row['away']
    homeTeam = league_table.query("Team == @homeName")
    awayTeam = league_table.query("Team == @awayName")
    hPos = int(homeTeam['Pos'])
    hGP = int(homeTeam['GP'])
    hGS = int(homeTeam['GS'])
    hGC = int(homeTeam['GC'])
    hCS = int(homeTeam['CS'])
    hPts = int(homeTeam['Pts'])
    aPos = int(awayTeam['Pos'])
    aGP = int(awayTeam['GP'])
    aGS = int(awayTeam['GS'])
    aGC = int(awayTeam['GC'])
    aCS = int(awayTeam['CS'])
    aPts = int(awayTeam['Pts'])
    
    if hGP == 0:
        h = [0,0,0,0]
    else:
        h = [round(x/hGP,2) for x in [hGS,hGC,hCS,hPts]]
    if aGP == 0:
        a = [0,0,0,0]
    else:
        a = [round(x/aGP,2) for x in [aGS,aGC,aCS,aPts]]
    
    stats = {'hPos':hPos,'hGSPG':h[0],'hGCPG':h[1],'hCSPG':h[2],'hPtsPG':h[3],'aPos':aPos,'aGSPG':a[0],'aGCPG':a[1],'aCSPG':a[2],'aPtsPG':a[3]}
    return stats
```

```python
# Now need a function to put this all together for a df.apply() method
def get_league_table_stats(row,df,teams):
    date = str(row['date'])
    season = str(row['season'])
    df_season = df.query("season == @season")
    league_table = basic_league_table(date,df_season,teams)
    data = get_useful_stats(league_table,row)
    #print(date)
    return data
```

```python
df = pd.read_csv("data_collected_final_version.csv")
df_new = df.copy()
df_new_26 = df_new.query("season == 26")
teams = df_new_26['home'].unique().tolist()
```

```python
t1 = time.time()
df_new_26[['hPos','hGSPG','hGCPG','hCSPG','hPtsPG','aPos','aGSPG','aGCPG','aCSPG','aPtsPG']] = df_new_26.apply(lambda row: pd.Series(get_league_table_stats(row,df_new_26,teams)),axis=1)
t2 = time.time()
print("Took {} seconds".format(round(t2-t1,5)))
```

```python
# Now we need to generalise get_league_table_stats to work for all seasons
def get_league_table_stats_all(row,df):
    date = str(row['date'])
    season = str(row['season'])
    df_season = df.query("season == @season")
    teams = df_season['home'].unique().tolist()
    league_table = basic_league_table(date,df_season,teams)
    data = get_useful_stats(league_table,row)
    return data
```

```python
df2 = df.copy()
t1 = time.time()
df2[['hPos','hGSPG','hGCPG','hCSPG','hPtsPG','aPos','aGSPG','aGCPG','aCSPG','aPtsPG']] = df2.apply(lambda row: pd.Series(get_league_table_stats_all(row,df2)),axis=1)
t2 = time.time()
print("Took {} seconds".format(round(t2-t1,5)))
```
