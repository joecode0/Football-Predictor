```python
# NEED TO TEST

def generate_league_table(df,start_date,end_date,output_file):
    df['Date'] = pd.to_datetime(df.Date)
    df.sort_values(by='Date',inplace=True,ascending=True)
    
    mask = (df['Date'] > start_date) & (df['Date'] <= end_date)
    subset_df = df.loc[mask]
    
    league_table_dict = {}
    teams = df['Home'].unique().tolist().sort()
    for i in range(20):
        league_table_dict[teams[i]] = {"W":0,"L":0,"D":0,"GS":0,"GC":0}
         
    # Need to add: POS, MP, GD, Pts
    
    for j in range(len(subset_df)):
        league_table_dict = update_table_with_result(league_table_dict,subset_df.iloc[j])
        
    league_table_dict = add_MP_GD_Pts(league_table_dict)
    
    final_league_table_dict = reorder_league_table(league_table_dict)
      
    return final_league_table_dict

def reorder_league_table(league_table_dict):
    points_dict = {}
    for team in league_table_dict.keys():
        points = league_table_dict[team].get('Pts')
        points_dict[team] = points
    league_table = {}
    i = 20
    for (team, _) in sorted(points_dict.iteritems(), key=lambda (k,v): (v,k)):
        teamData = league_table_dict.get(team)
        teamData['Pos'] = i
        league_table[team] = teamData
        i -= 1
    return league_table

def add_MP_GD_Pts(league_table_dict):
    for team in league_table_dict.keys():
        dataRow = league_table_dict.get(team)
        dataRow['Pts'] = 3*dataRow['W'] + dataRow['D']
        dataRow['MP'] = dataRow['W'] + dataRow['D'] + dataRow['L']
        dataRow['GD'] = dataRow['GS'] - dataRow['GC']
        league_table_dict[team] = dataRow
    return league_table_dict


def update_table_with_result(league_table_dict,row):
    homeTeam = str(row['Home'])
    awayTeam = str(row['Away'])
    result = str(row['Result'])
    [homeGoals,awayGoals] = result.split("-")
    hg = int(homeGoals)
    ag = int(awayGoals)
    
    homeRow = league_table_dict.get(homeTeam)
    homeRow["GS"] += hg
    homeRow["GC"] += ag
    
    awayRow = league_table_dict.get(awayTeam)
    awayRow["GS"] += ag
    awayRow["GC"] += hg
    if hg > ag:
        homeRow["W"] += 1
        awayRow["L"] += 1
    elif hg == ag:
        homeRow["D"] += 1
        awayRow["D"] += 1
    else:
        homeRow["L"] += 1
        awayRow["W"] += 1
    
    league_table_dict[homeTeam] = homeRow
    league_table_dict[awayTeam] = awayRow
    return league_table_dict
        
    
```
