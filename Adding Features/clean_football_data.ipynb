```python
import pandas as pd
```

```python
df = pd.read_csv("football-data-raw.csv")
```

```python
df.head()
```

```python
str(pd.to_datetime(df.iloc[0].Date,format='%d/%m/%y'))[0:10]
```

```python
def fix_date_column(row):
    date = str(row['date'])
    try:
        return str(pd.to_datetime(date,format='%d/%m/%y'))[0:10]
    except ValueError:
        return str(pd.to_datetime(date,format='%d/%m/%Y'))[0:10]

def fix_team_column(row,col):
    team = row[col]
    if team == "Sheffield United":
        return "sheffield utd"
    elif team == "Man United":
        return "man utd"
    elif team == "Tottenham":
        return "spurs"
    elif team == "Nott'm Forest":
        return "nottingham forest"
    elif team == "Middlesboro":
        return "middlesbrough"
    else:
        return str(team).lower()
```

```python
df.columns = [str(x).lower() for x in list(df.columns.values)]
```

```python
df['date'] = df.apply(fix_date_column,axis=1)
```

```python
df['hometeam'] = df.apply(fix_team_column,args=('hometeam',),axis=1)
```

```python
df['awayteam'] = df.apply(fix_team_column,args=('awayteam',),axis=1)
```

```python
df.head()
```

```python
df.sort_values(by=['date','hometeam'],inplace=True)
```

```python
df.shape
```

```python
df.dropna(inplace=True,subset=['fthg'])
```

```python
df.shape
```

```python
df.fillna(0,inplace=True)
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
df.to_csv("football-data-cleaned.csv",index=False)
```
