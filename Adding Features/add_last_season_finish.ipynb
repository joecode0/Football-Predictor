```python
import pandas as pd
df = pd.read_csv("main_features_v2.csv")
%run league-table-generator-copy.ipynb

def create_all_final_league_tables(df):
    seasons = df['season'].unique().tolist()
    tables = {}
    for s in seasons:
        df_season = df.query("season == @s")
        date = pd.to_datetime('now')
        teams = df_season['home'].unique().tolist()
        table = basic_league_table(date,df_season,teams)
        tables[s] = table
    return tables
```

```python
df2 = pd.read_csv("main_database_v0.csv")
league_tables = create_all_final_league_tables(df2)
```

```python
def get_pos(team,table):
    if team in table['Team'].unique().tolist():
        entry = table.query("Team == @team")
        return int(entry['Pos'])
    else:
        return int(18)
```

```python
def get_last_season_pos(row,homeBool,tables):
    if homeBool:
        team = row['home']
    else:
        team = row['away']
        
    season = row['season']
    table = tables[int(season)-1]
    pos = get_pos(team,table)
    return pos
```

```python
df['hPosLS'] = df.apply(get_last_season_pos,args=(True,league_tables),axis=1)
df['aPosLS'] = df.apply(get_last_season_pos,args=(False,league_tables),axis=1)
```
