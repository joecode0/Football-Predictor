```python
import pandas as pd
```

```python
df_main = pd.read_csv("main_database_v0.csv")
df_fd = pd.read_csv("football-data-cleaned.csv")
```

```python
df_main.head()
```

```python
df_fd.head()
```

```python
def valid_date(value,compare_date):
    date = pd.to_datetime(value)
    if date < compare_date:
        return False
    else:
        return True
```

```python
df2 = df_fd[[valid_date(x,pd.to_datetime("2000-08-18")) for x in df_fd.date.tolist()]]
```

```python
df1 = df_main[[valid_date(x,pd.to_datetime("2000-08-18")) for x in df_main.date.tolist()]]
```

```python
df1.reset_index(drop=True,inplace=True)
df2.reset_index(drop=True,inplace=True)
```

```python
cols = list(df2.columns.values)
cols[1] = 'home'
cols[2] = 'away'
cols[6] = 'ref'
cols[0:7]
```

```python
df2.columns = [str(x) + "2" for x in list(df2.columns.values)]
```

```python
df2.head()
```

```python
def get_score(row):
    home_goals = str(int(row['fthg2']))
    away_goals = str(int(row['ftag2']))
    return home_goals + "-" + away_goals
```

```python
df2['score2'] = df2.apply(get_score,axis=1)
```
