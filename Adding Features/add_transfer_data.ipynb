```python
def fix_money_col(row,col):
    money = str(row[col])
    if money[-1] == "m":
        return str(money[1:-1])
    elif money[-1] == "k":
        return str("0."+str(money[1:-1]))
    else:
        return "0"

transfer_data['expenditure'] = transfer_data.apply(fix_money_col,args=('Expenditure',),axis=1)  
```

```python
transfer_data['income'] = transfer_data.apply(fix_money_col,args=('Income',),axis=1)
```

```python
transfer_data.head()
```

```python
transfer_data.drop(['Expenditure','Income'],inplace=True,axis=1)
```

```python
def get_net_income(row):
    exp = float(row['expenditure'])
    income = float(row['income'])
    net = round(float(income-exp),3)
    return net

transfer_data['net'] = transfer_data.apply(get_net_income,axis=1)
```

```python
transfer = transfer_data.query("Year > 1999")
```

```python
transfer.columns = ['year','team','spent','income','net']
```

```python
transfer.tail()
```

```python
def year_to_season(row):
    year = row['year']
    season = 9 + (year-2000)
    return season
transfer['season'] = transfer.apply(year_to_season,axis=1)
transfer.drop(['year'],axis=1,inplace=True)
transfer = transfer[['season','team','spent','income','net']]
transfer.head()
```

```python
def get_transfer_data(row,transfer,homeBool):
    season = row['season']
    relevant_transfer_data = transfer.query("season == @season")
    if homeBool:
        team = row['home']
    else:
        team = row['away']
    data = relevant_transfer_data.query("team == @team")
    spent = float(data['spent'])
    income = float(data['income'])
    net = float(data['net'])
    return spent,income,net

df['hSpent'], df['hIncome'], df['hNet'] = zip(*df.apply(get_transfer_data,args=(transfer,True),axis=1))
```

```python
df['aSpent'], df['aIncome'], df['aNet'] = zip(*df.apply(get_transfer_data,args=(transfer,False),axis=1))
```
