```python
import pandas as pd
```

```python
df = pd.read_csv("main_with_shots.csv")
```

```python
df.head()
```

```python
def shots_stats_so_far(row,df,homeBool):
    date = row['date']
    season = row['season']
    df_season = df.query('season == @season')
    games_so_far = df_season[pd.to_datetime(df.date) < pd.to_datetime(date)]
    
    if homeBool:
        team = row['home']
    else:
        team = row['away']
    games_when_home = games_so_far.query("home == @team")
    games_when_away = games_so_far.query("away == @team")
    
    home_shots = sum(games_when_home['hs2'].tolist())
    home_shots_on_target = sum(games_when_home['hst2'].tolist())
    away_shots = sum(games_when_away['as2'].tolist())
    away_shots_on_target = sum(games_when_away['ast2'].tolist())
    total_games = len(games_when_home) + len(games_when_away)
    
    if total_games == 0:
        return 0,0
    
    shots_per_game = round((home_shots + away_shots)/total_games,3)
    on_target_per_game = round((home_shots_on_target + away_shots_on_target)/total_games,3)
    
    return shots_per_game, on_target_per_game
```

```python
#df[['hSPG','hSTPG']] = df.apply(shots_stats_so_far,args=(df,True),axis=1,result_type='broadcast')
df['hSPG'], df['hSTPG'] = zip(*df.apply(shots_stats_so_far,args=(df,True),axis=1))
df['aSPG'], df['aSTPG'] = zip(*df.apply(shots_stats_so_far,args=(df,False),axis=1))
```

```python
def shooting_percentage(row,homeBool):
    if homeBool:
        SPG = int(row['hSPG'])
        GPG = int(row['hGSPG'])
    else:
        SPG = int(row['aSPG'])
        GPG = int(row['aGSPG'])
    if SPG == 0:
        return 0
    else:
        return round(GPG/SPG,3)
```

```python
df['hS%'] = df.apply(shooting_percentage,args=(True,),axis=1)
df['aS%'] = df.apply(shooting_percentage,args=(False,),axis=1)
```
