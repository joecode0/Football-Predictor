```python
import pandas as pd
df1 = pd.read_csv("main_2000_onwards.csv")
df2 = pd.read_csv("football_2000_onwards.csv")
df1.shape
```

```python
df = pd.concat([df1,df2],axis=1)
```

```python
df.head()
```

```python
def cols_agree(row,col1,col2):
    return (str(row[col1]) == str(row[col2]))
```

```python
df['home_agree'] = df.apply(cols_agree,args=('home','home2'),axis=1)
df['away_agree'] = df.apply(cols_agree,args=('away','away2'),axis=1)
df['score_agree'] = df.apply(cols_agree,args=('score','score2'),axis=1)
df['date_agree'] = df.apply(cols_agree,args=('date','date2'),axis=1)
```
